```python
import pandas as pd
# Load dataset
df = pd.read_csv('abalone.csv')
```

```python
# Show first rows
print(df.head())
```

```python
# Basic info
print(df.info())
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
# Summary statistics
print(df.describe())
```

```python
import matplotlib.pyplot as plt
#Visualization 1: Rings Distribution (Age)
plt.figure()
df['Rings'].plot(kind='hist', bins=20)
plt.title('Distribution of Rings (Age)')
plt.xlabel('Rings')
plt.show()
```

```python
# Visualization 2: Average Rings by Sex
avg_rings = df.groupby('Sex')['Rings'].mean()

plt.figure()
avg_rings.plot(kind='bar')
plt.title('Average Age (Rings) by Sex')
plt.xlabel('Sex')
plt.ylabel('Average Rings')
plt.show()
```

```python
# Visualization 3: Weight vs Rings
plt.figure()
plt.scatter(df['Whole weight'], df['Rings'])
plt.title('Whole Weight vs Rings')
plt.xlabel('Whole Weight')
plt.ylabel('Rings')
plt.show()
```
